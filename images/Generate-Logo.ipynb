```python
from PIL import Image, ImageFont, ImageDraw

width = 1280
height = 640
half_height = height // 2

icon_width = 320
icon_height = 320
icon_size = (icon_width, icon_height)

font1 = ImageFont.truetype("BRLNSDB.TTF", 96)
font2 = ImageFont.truetype("DUBAI-MEDIUM.TTF", 96)

font1_color = (25, 25, 25, 255)
bot_color = (171, 120, 228, 255)
web_color = (61, 187, 248, 255)

bgColor = (240, 240, 240, 255)

rokuga_chan = Image.open("bot-icon.png").resize(icon_size)
shichou_chan = Image.open("web-icon.png").resize(icon_size)

line_width = 10

img = None
img = Image.new("RGBA", (width, height))
draw = ImageDraw.Draw(img)

ellipse_offset = 24
ellipse_bound = [(8, 0), (320 + 8 + 4, 320 + 4)]
draw.ellipse(ellipse_bound, fill=(0,0,0,255))

img.paste(rokuga_chan, box=(10, 0))
img.paste(shichou_chan, box=(img.width - shichou_chan.width, half_height))
# draw.rectangle(ellipse_bound, outline=(0, 0, 255, 255))

name_offset_x = 30
name_offset_y = 40

word_offset_x = 100
word_offset_y = 140

shichou_offset_x = 220

draw.text((icon_width + name_offset_x, name_offset_y), "Rokuga-Chan", font=font1, fill=font1_color)
draw.text((icon_width + word_offset_x, word_offset_y), "Recording for you.", font=font2, fill=bot_color)

draw.text((shichou_offset_x + name_offset_x, half_height + name_offset_y), "Shichou-Chan", font=font1, fill=font1_color)
draw.text((shichou_offset_x + word_offset_x, half_height + word_offset_y), "Watch with you.", font=font2, fill=web_color)

draw.line([(0, half_height), (width, half_height)], width=line_width, fill=font1_color)

img
```
